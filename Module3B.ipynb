# Module 3B
### Module 3 covers generating spharm-obj organelle models, creating reports from them, and comparing them.
### Module 3B generates two reports, each of which compares models of two different organelles

For more details, please refer to our [paper](https://doi.org/10.1093/bioinformatics/btac688): Improving and evaluating deep learning models of cellular organization. Huangqingbo Sun,  Xuecong Fu,  Serena Abraham,  Shen Jin,  Robert F Murphy. Bioinformatics, Volume 38, Issue 23, 1 December 2022, Pages 5299â€“5306.

```python
# Code Cell 1

from cellorganizer.tools import slml2report
import glob
import urllib.request
import tarfile 
import os
import scipy.io
from IPython.core.display import display, HTML, Image

topdir = "/home/murphylab/cellorganizer/local/results"
folder_name = 'Module3B'
outputdir = topdir + "/" + folder_name
if not os.path.exists(topdir):
    os.makedirs(topdir)
os.chdir(topdir)
if not os.path.exists(outputdir):
    os.makedirs(outputdir)
os.chdir(outputdir)

model_name = 'golgi_model_vs_nuclear_model'
if not os.path.exists(model_name):
    os.makedirs(model_name)
os.chdir(model_name)

options = {}
options['verbose'] = 1
options['includenuclear'] = 0
options['includecell'] = 0
options['includeprot'] = 1
options['paired'] = 0
```

* Compare the golgi appratus model with the nucleoli model

```python
# Code Cell 2

model_folder = "/home/murphylab/cellorganizer/local/models/"
model_files_1 = model_folder + 'golgi_GT_model_s.mat'
model_files_2 = model_folder + 'nucleoli_GT_model_s.mat'
os.system("ls")
answer = slml2report(model_files_1 , model_files_2, options)
#save the scores from this comparison
#GolgiNucleoliComparison = scipy.io.loadmat('spharm_obj_comparison.mat')
HTML(filename =  'index.html')
```
